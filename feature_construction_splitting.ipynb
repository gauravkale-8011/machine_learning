# 1) Feature Construction

```python
df.head()
```

```python
df["Family_size"]=df["SibSp"]+df["Parch"]+1
```

```python
df.head()
```

```python
def family_type(size):
    if size == 1:
        return 0
        #small family
    if size > 1 and size <=4:
        return 1
        #medium family
    else:
        return 2
```

```python
family_type(1)
```

```python
df["Family_type"]=df["Family_size"].apply(family_type)
```

```python
df.head()
```

```python
new_df=df.drop(columns=["SibSp","Parch"])
```

```python
new_df.head()
```

```python
x=new_df.drop(columns=["Survived"])
y=new_df["Survived"]
```

```python
print("accuracy",np.mean(cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=20)))
# so basically here we can see that the accuracy is improved
```
