### Ordinal encoding

* if the input column is ordinal then we do ordinal encoding

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:2],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
x_train
```

- so basically here we have to  assign the ordinal column like this we have done in object creating process
- always understand while transforming first do fit() and only train the model on train data

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
oe.fit(x_train)

x_train_encode=oe.transform(x_train)
x_test_encode=oe.transform(x_test)
```

- so basically we have done here ordinal encoding 
- so we do ordinal encoding on input columns if the feature is categorical(>>ordinal)
- ordinal ex:- good>avg>bad

### Label encoding

- when the output column is categorical then we use label encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

le.fit(y_train)

y_train_encode=le.transform(y_train)
y_test_encode=le.transform(y_test)
```

### One Hot Encoding

```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cars.csv")
df
```

# a) one hot endocing using pandas

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:4],df.iloc[:,-1],test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder(drop="first",dtype=np.int32)

x_train_encode=oh.fit_transform(x_train[["fuel","owner"]]).toarray()
x_test_encode=oh.transform(x_test[["fuel","owner"]]).toarray()
```
