```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("covid_toy.csv")
df.head()
```

### 1) Aam Zindagi

* performing simple imputer on column fever because some values are not present in that column

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:5],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()

x_train_fever=si.fit_transform(x_train[["fever"]])
x_test_fever=si.transform(x_test[["fever"]])
```

* performing ordinal encoder on cough column because it is ordinal categorical data

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[["Mild","Strong"]])

x_train_cough=oe.fit_transform(x_train[["cough"]])
x_test_cough=oe.transform(x_test[["cough"]])
```

* peerforming onehotencoding on clumn gender and city beacause they are nominal categorical column

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop="first",dtype=np.int32)

x_train_gc=ohe.fit_transform(x_train[["gender","city"]]).toarray()
x_test_gc=ohe.transform(x_test[["gender","city"]]).toarray()
```

```python
x_train_age=x_train[["age"]].values
x_test_age=x_test[["age"]].values
```

```python
x_train_transform=np.hstack((x_train_fever,x_train_cough,x_train_gc,x_train_age))
```

```python
x_train_transform.shape
```

### 2) Mentos Zindagi

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ("tnf1",SimpleImputer(),["fever"]),
    ("tnf2",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
    ("tnf3",OneHotEncoder(drop="first",sparse_output=False),["gender","city"])
],
remainder="passthrough")
```

```python
x_train_transform=transformer.fit_transform(x_train)
x_test_transform=transformer.transform(x_test)
```

```python
x_train_transform.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv("customer.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:4],df.iloc[:,-1],test_size=0.3,random_state=0)
```

```python
x_train_transform=transformer.fit_transform(x_train)
x_test_transform=transformer.fit_transform(x_test)
```
