```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic (1).csv")
df.head()
```

```python
df["number"].unique()
```

```python
df["number"].value_counts().plot(kind="bar")
```

- here to_numeric take numbers from data and make other nan like string wii be denoted as string
- np.where(condition, if_true, if_false)

```python
df["number_numerical"]=pd.to_numeric(df["number"],errors="coerce",downcast="integer")
df["number_categorical"]=np.where(df["number_numerical"].isnull(),df["number"],np.nan)
```

```python
df.head()
```

```python
df["Cabin"].unique()
```

```python
df["Ticket"].unique()
```

```python
df["Cabin_numerical"]=df["Cabin"].str.extract("(\d+)") # extract number
```

```python
df["Cabin Categorical"]=df["Cabin"].str[0] # extract first
```

```python
df.head()
```

```python
df['Cabin Categorical'].value_counts().plot(kind='bar')
```

```python
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df["ticket_num"]=pd.to_numeric(df["ticket_num"],errors="coerce",downcast="integer")

df["ticket_cat"]=df["Ticket"].apply(lambda s : s.split()[0])
df["ticket_cat"]=np.where(df["ticket_cat"].str.isdigit(),np.nan,df["ticket_cat"])
```
