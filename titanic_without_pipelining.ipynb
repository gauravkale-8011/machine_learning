```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=["PassengerId","Name","Ticket","Cabin"],inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder 
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]],df["Survived"],test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
df.isnull().sum()
```

```python
# imputing missing values
age_impute=SimpleImputer()
emb_impute=SimpleImputer(strategy="most_frequent")

age_train=age_impute.fit_transform(x_train[["Age"]])
emb_train=emb_impute.fit_transform(x_train[["Embarked"]])

age_test=age_impute.transform(x_test[["Age"]])
emb_test=emb_impute.transform(x_test[["Embarked"]])                                 
```

```python
ohe_sex = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
ohe_emb = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# TRAIN
sex_train = ohe_sex.fit_transform(x_train[["Sex"]])
Embarked_train = ohe_emb.fit_transform(emb_train)

# TEST
sex_test = ohe_sex.transform(x_test[["Sex"]])
Embarked_test = ohe_emb.transform(emb_test)
```

```python
other_train=x_train[["Pclass","SibSp","Parch","Fare"]]
other_test=x_test[["Pclass","SibSp","Parch","Fare"]]
```

```python
x_train_transformed=np.concatenate((other_train,age_train,sex_train,Embarked_train),axis=1)
x_test_transformed=np.concatenate((other_test,age_test,sex_test,Embarked_test),axis=1)
```

```python
x_train_transformed
```

```python
from sklearn.tree import DecisionTreeClassifier
obj=DecisionTreeClassifier()

obj.fit(x_train_transformed,y_train)
```

```python
y_pred=obj.predict(x_test_transformed)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
