```python
import pandas as pd
import pickle
import numpy as np
```

```python
ohe_sex=pickle.load(open("models/ohe_sex.pkl","rb"))
ohe_emb=pickle.load(open("models/ohe_emb.pkl","rb"))
obj=pickle.load(open("models/obj.pkl","rb"))
```

```python
test_input = np.array([3, 'male', 35.0, 0,0,8.0500, 'S'],dtype=object).reshape(1,7)
```

```python
ohe_sex_transform=ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
ohe_emb_transform=ohe_emb.transform(test_input[:,6].reshape(1,1))
```

```python
age=test_input[:,2].reshape(1,1)
```

```python
remaining=test_input[:,[0,3,4,5]]
```

```python
all_transform=np.concatenate((remaining,age,ohe_sex_transform,ohe_emb_transform),axis=1)
```

```python
obj.predict(all_transform)
```
