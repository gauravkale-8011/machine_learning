```python
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.impute import SimpleImputer,KNNImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("train.csv",usecols=["Survived","Age","Pclass","Fare"])
df.head()
```

```python
df.isnull().mean()*100
```

```python
x=df.drop(columns=["Survived"])
y=df["Survived"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
kn=KNNImputer(n_neighbors=5,weights="distance")
kn.fit(x_train)
# so basically here we have ally knn imputer in which we have pass the value of k means it will take the mean of that much nearest value
# weight will distance or uniform by default it is uniform means normal mean but distance means multplying each value with the recprocal of the distance from that point
```

```python
x_train_transform=kn.transform(x_train)
x_test_transform=kn.transform(x_test)
```

```python
x_train_transform
```

```python
lt=LogisticRegression()
lt.fit(x_train_transform,y_train)
```

```python
y_pred=lt.predict(x_test_transform)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
# now we will see without knn accuracy
```

```python
st=SimpleImputer()
st.fit(x_train)
x_train_trf=st.transform(x_train)
x_test_trf=st.transform(x_test)
```

```python
lt=LogisticRegression()
lt.fit(x_train_trf,y_train)
```

```python
y_pred=lt.predict(x_test_trf)
```

```python
accuracy_score(y_test,y_pred)
```
